# Talk2Mind - Exploratory Data Analysis
Run `data/generate_dataset.py` first to produce the CSV files this notebook reads.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/talk2mind_dataset.csv')
df.head()
```

```python
df['risk_category'].value_counts().plot(kind='bar', title='Risk Category Distribution')
plt.ylabel('Count')
plt.show()
```

```python
df['mental_wellbeing_score'].plot(kind='hist', bins=30, title='Mental Well-Being Score Distribution')
plt.xlabel('Score (0-100)')
plt.show()
```

```python
import numpy as np
corr_cols = ['phq9_total', 'gad7_total', 'facial_muscle_tension', 'pause_ratio', 'mental_wellbeing_score']
df[corr_cols].corr()
```

```python
df.groupby('occupation')['mental_wellbeing_score'].mean().plot(kind='bar', title='Avg Well-Being Score by Occupation')
plt.ylabel('Avg Score')
plt.show()
```
